```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('../data/pallet_test/pallet_model.xyz', delimiter=' ')
```

```python
data.shape
```

```python
# plot 3d coordinates, do not scale axes (aspect is equal in all directions)
print(data[:, 2].min(), data[:, 2].max())
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:, 0], data[:, 1], data[:, 2])
ax.set_aspect('equal')
plt.show()
```

```python
data_fixed = data.copy()
data_fixed = data_fixed[data_fixed[:, 2] > 0, :]
data_fixed = data_fixed[(data_fixed[:, 0] < 99) & (data_fixed[:, 0] > 60) & (data_fixed[:, 1] < 10), :]
# data_fixed = data_fixed[(data_fixed[:, 0] < 5) | (data_fixed[:, 0] > 90) | (data_fixed[:, 1] < 5) | (data_fixed[:, 1] > 100), :]
# save to file in xyz format with 3 decimal places
np.savetxt('../data/pallet_test/box_packing_model_fixed.xyz', data_fixed, fmt='%.3f', delimiter=' ')
```

```python
# plot 3d coordinates, do not scale axes (aspect is equal in all directions)
print(data_fixed[:, 2].min(), data_fixed[:, 2].max())
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_fixed[:, 0], data_fixed[:, 1], data_fixed[:, 2])
ax.set_aspect('equal')
plt.show()
```

```python
print(data_fixed[:, 0].min(), " ", data_fixed[:, 0].max())
print(data_fixed[:, 1].min(), " ", data_fixed[:, 1].max())
print(data_fixed[:, 2].min(), " ", data_fixed[:, 2].max())
```
