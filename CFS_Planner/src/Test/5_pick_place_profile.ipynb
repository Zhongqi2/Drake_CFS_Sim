```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
optim_path = '/home/ruic/Documents/MASCEI/Auto_calibration/optimized_traj.txt'
data = np.loadtxt(optim_path)

# Create a figure with six subplots
fig, axs = plt.subplots(6, 1, figsize=(8, 8), sharex=True)

# Plot the path in each subplot
for i in range(6):
    axs[i].plot(data[i])  # Replace x_values and y_values with your actual data
    axs[i].set_title(f'Joint {i+1} Position (rad)')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()

fig.savefig('optimized_pos.png')
```

```python
dt = 0.004
vel = (data[:, 1:] - data[:, :-1]) / dt

# Create a figure with six subplots
fig, axs = plt.subplots(6, 1, figsize=(8, 8), sharex=True)

# Plot the path in each subplot
for i in range(6):
    axs[i].plot(np.arange(vel.shape[1])*dt, vel[i])  # Replace x_values and y_values with your actual data
    axs[i].set_title(f'Joint {i+1} Velocity (rad/s)')
    axs[i].plot([0.0*dt, vel.shape[1]*dt], [5.0, 5.0], 'r--')
    axs[i].plot([0.0*dt, vel.shape[1]*dt], [-5.0, -5.0], 'r--')

axs[-1].set_xlabel('Time (s)')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()

fig.savefig('optimized_vel.png')
```
